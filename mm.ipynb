```python
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
from PIL import Image
import io
import lxml

query = 'memes'
# Make a request to the website
url = f'https://www.bing.com/images/search?cw=1465&ch=714&q={query}&first=1'  # Replace with the actual URL
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.content, 'lxml')

# Find all image tags in the HTML
image_tags = soup.find_all('img')

image_urls = []
for img in image_tags:
    # print(img)
    if 'src' in img.attrs :
        src = img['src']
        if src.startswith('https://'):
            image_urls.append(src.replace('&w=42&h=42',''))
        # image_urls.append(img['src'].replace('&w=42&h=42',''))
    if 'src2' in img.attrs:
        src = img['src2']
        if src.startswith('https://'):
            image_urls.append(src.replace('&w=42&h=42',''))
    if 'data-src' in img.attrs:
        src = img['data-src']
        if src.startswith('https://'):
            image_urls.append(src.replace('&w=42&h=42',''))


# image_urls


```

```python
# Iterate over the first 16 images (4x4 grid)
for i in range(20):
    # Make a request to the image URL
    img_response = requests.get(image_urls[i])
    img_data = img_response.content

    # Create a BytesIO object to read the image data
    img_buffer = io.BytesIO(img_data)

    # Open the image using PIL (Python Imaging Library)
    img = Image.open(img_buffer)

    # Display the image
    plt.subplot(4, 6, i + 1)
    plt.imshow(img)
    plt.axis('on')

# Show the grid of images
plt.tight_layout()
plt.show()
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import json

# Wikipedia page URL with the table
url = 'https://en.wikipedia.org/wiki/List_of_states_and_union_territories_of_India_by_population'  # Replace with the actual Wikipedia page URL

# Send a GET request to the URL and retrieve the HTML content
response = requests.get(url)

# response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table element on the page (modify the selector as needed)
table = soup.find_all('table')

# Extract the table data into a list of dictionaries
table_data = []
header_row = table.find('tr')  # Assumes the header row is the first row of the table
headers = [header.text.strip() for header in header_row.find_all('th')]
for row in table.find_all('tr')[1:]:  # Start from the second row to exclude the header row
    row_data = {}
    cells = row.find_all('td')
    for i, cell in enumerate(cells):
        row_data[headers[i]] = cell.text.strip()
    table_data.append(row_data)

# Save the table data as JSON
with open('table_data.json', 'w') as f:
    json.dump(table_data, f)
```
